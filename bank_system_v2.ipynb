```python
import textwrap
```

```python
def menu():
  menu = """\n
  =========== Menu ===========
  [1]\tDepositar
  [2]\tSacar
  [3]\tGerar Extrato
  [4]\tCriar Usuário
  [5]\tCriar Conta
  [6]\tListar Contas
  [7]\tSair\n
  """

  return int(input(textwrap.dedent(menu)))
```

```python
def depositar(saldo, valor, extrato, /):  # tudo que está antes do / é posicional, i.e., não aceita valor=valor etc.
  if valor > 0:
    saldo += valor
    extrato += f"Depósito no valor de : R$ {valor:.2f}\n"
    print("\nDepósito efetuado com sucesso!")

  else:
    print("Operação não efetuada. Valor informado não pode ser nulo ou negativo")

  return saldo, extrato
```

```python
def sacar(*, saldo, valor, extrato, limite, numero_saques, limite_saques):  # após o asterisco tudo é nomeado!
  excedeu_saldo = valor > saldo
  excedeu_limite = valor > limite
  excedeu_saques = numero_saques > limite_saques

  if excedeu_saldo:
    print("Saldo insuficiente")

  elif excedeu_limite:
    print("Operação acima do limite de saque. Seu limite de saque atual é de R$ 500,00")

  elif excedeu_saques:
    print("Limite de saques no dia atingido. Aguarde até o dia seguinte ou fale com seu gerente")

  elif valor > 0:
    saldo -= valor
    extrato += f"Saque no valor de: R$ {valor:.2f}\n"
    numero_saques += 1
    print("Saque realizado com sucesso!")

  else:
    print("Valor informado é inválido! Informe um valor não nulo")

  return saldo, extrato
```

```python
def exibir_extrato(saldo, /, *, extrato):  # saldo é posicional only e extrato é nomeado only
  print("\n=========  Extrato  =========")
  print("Sem transações registradas" if not extrato else extrato)  # parece list comprehension
  print(f"\nSaldo:\t\tR$ {saldo:.2f}")  # \t\t == dois recuos-padrão
  print("\n=============================")
```

```python
def criar_usuario(usuarios):
  cpf = input("Informe o CPF, somente números: ")
  usuario = filtrar_usuario(cpf, usuarios)

  if usuario:
    print("Já existe um usuário registrado com esse cpf")
    return

  nome = input("Informe o nome completo: ")
  data_nascimento = input("Informe data de nascimento, no modelo dd/mm/aaaa: ")
  endereco = input("Informe o endereço conforme ordem: logradouro, nº - bairro - cidade/sigla estado: ")

  usuarios.append({"nome": nome, "data_nasc": data_nascimento, "cpf": cpf, "endereco": endereco})

  print("Usuário cadastrado com sucesso!")
```

```python
def filtrar_usuario(cpf, usuarios):
  usuarios_filtrados = [usuario for usuario in usuarios if usuario["cpf"] == cpf]

  return usuarios_filtrados[0] if usuarios_filtrados else None
```

```python
def criar_conta(agencia, numero_conta, usuarios):
  cpf = input("Informe o CPF do usuário: ")
  usuario = filtrar_usuario(cpf, usuarios)

  if usuario:
    print("Conta criada com sucesso!")
    return {"agencia": agencia, "numero_conta": numero_conta, "usuario": usuario}

  print("Usuário não encontrado, encerrando criação de conta")
```

```python
def listar_contas(contas):
  for conta in contas:
    linha = f"""\
        Agência:\t{conta["agencia"]}
        C/C:\t\t{conta["numero_conta"]}
        Titular:\t{conta["usuario"]["nome"]}
        """
    print("=" * 50)
    print(textwrap.dedent(linha))
```

```python
def main():
  LIMITE_SAQUES = 3
  AGENCIA = "0001"
  LIMITE_SAQUE_VALOR = 500

  saldo = 0
  extrato = ""
  numero_saques = 0
  usuarios = []
  contas = []

  while True:
    op = menu()

    if op == 1:  # DEPÓSITO
      valor = float(input("Operação de Depósito selecionada.\nIndique o valor: "))

      saldo, extrato = depositar(saldo, valor, extrato)

    elif op == 2:  # SAQUE
      valor = float(input("Operação de Saque selecionada.\nIndique o valor: "))

      saldo, extrato = sacar(
          saldo=saldo,
          valor=valor,
          extrato=extrato,
          limite=LIMITE_SAQUE_VALOR,
          numero_saques=LIMITE_SAQUES,
      )

    elif op == 3:  # EXTRATO
      exibir_extrato(saldo, extrato=extrato)  # saldo como posicional e extrato como nomeado

    elif op == 4:  # CRIAR USUÁRIO
      criar_usuario(usuarios)

    elif op == 5:  # CRIAR CONTA
      numero_conta = len(contas) + 1
      conta = criar_conta(AGENCIA, numero_conta, usuarios)

      if conta:
        contas.append(conta)

    elif op == 6:  # LISTAR CONTAS
      listar_contas(contas)

    elif op == 7:  # SAIR
      break

    else:
      print("Operação inválida, selecione novamente a operação desejada ou entre em contato com seu gerente")
```

```python
main()
```
